1. **¿Qué estructura de estado utilizarían?**

Como estructura de estados utilizaríamos una matriz 3x3. Esto es porque permite representar fácilmente las posiciones de los números además de facilitar el movimiento entre filas y columnas en el código.

2. **Al menos 2 heurísticas admisibles no-triviales**

Las dos heurísticas que elegimos son la **Distancia de Manhattan** y **Conflicto Lineal + Manhattan**

**Distancia de Manhattan:**

Calcula la suma de las distancias horizontales y verticales que cada ficha debe recorrer desde su posición actual hasta su posición objetivo. Formalmente, para cada ficha `i`, se computa:

$$
h_1(n)=\sum^8_{i=1}(|x_i-x_i^*|+|y_i-y_i^*)
$$

donde $(x_i,y_i)$ es la posición actual de la ficha y $(x_i^*,y_i^*)$ es su posición objetivo.
